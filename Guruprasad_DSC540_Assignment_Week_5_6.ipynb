### 7.4 Transform data using either mapping or a function

```python
# In this section the Country field is cleansed and a new column for "Clean Country name is added"
vc_country=candy_df['Q4: COUNTRY'].value_counts()
vc_country.head(5)
```

```python
# Creating a filter with possible options for countries with name United states
vc_country_usa=vc_country[((vc_country.index.str.contains('US')) | (vc_country.index.str.contains('UNITED')) | (vc_country.index.str.contains('AMERICA')) ) & (~vc_country.index.isin(['UNITED KINGDOM','AUSTRALIA']))].index
# Creating a filter with possible options for countries with name Canada
vc_country_canada=vc_country[(vc_country.index.str.contains('CANADA'))].index
# Creating a filter with possible options for countries with name United Kingdom
vc_country_uk=vc_country[vc_country.index.isin(['UNITED KINGDOM','UK'])].index
# Creating a filter with possible options for countries with name Korea
vc_country_korea=vc_country[(vc_country.index.str.contains('KOREA'))].index
```

```python
# Creating a dictionary of Clean country name with possible values from the step above
CLEAN_COUNTRY_NAME_dict = {"USA":vc_country_usa,
                     "CANADA":vc_country_canada,
                     "UK":vc_country_uk,
                     "KOREA":vc_country_korea}
# For other country names that does not require cleansing, they are added as is to the clean_country_dict
other_countries=["GERMANY","JAPAN","FRANCE","IRELAND","SCOTLAND","AUSTRALIA","COSTA RICA","UAE","ENGLAND","ICELAND","CHINA",
 "SPAIN","SINGAPORE","NETHERLANDS","MEXICO","DENMARK","INDIA"]
# THe countries in other_countries does not require cleansing, they are added to the dict as is.
for countr in other_countries:
    CLEAN_COUNTRY_NAME_dict[countr]=countr
CLEAN_COUNTRY_NAME_dict
```

```python
def get_country(val):
    """
    This function parses through the clean country dictionary and looks for the value passed to the function.
    If a Match is found it returns the cleansed Country Name
    """
    for key, vals in CLEAN_COUNTRY_NAME_dict.items():
        if val in vals:
            return(key)
        else:
            continue
            
# Using the Map function on the Q4: COUNTRY variable to clean the country name and added a new column            
candy_df['CLEAN_COUNTRY_NAME']=candy_df['Q4: COUNTRY'].map(get_country)
candy_df[['Q4: COUNTRY','CLEAN_COUNTRY_NAME']].head(10)
```

### 10.4. Split, Apply, Combine

```python
# Creating functions to be used in the apply function
def get_count(group):
    """
    This function will return the count of observations in each group
    """
    return group.count()

def desc(group):
    """
    This function will return the result of decribe function for the group
    """
    return group.describe()

def sort_within_group(group):
    """
    This function returns the top 10 results sorted in descending order in the group
    """
    return group.sort_values(ascending=False)[:10]
```

```python
# creating a groupby object which performs the split
candy_grouped3=candy_subset_df17.groupby(["CLEAN_COUNTRY_NAME","Q5: STATE, PROVINCE, COUNTY, ETC"])["Q3: AGE"]
```

```python
# Using the apply method on the function get_count which performs apply and combines the results
candy_grouped3.apply(get_count)
```

```python
# SOrting the counts within the group by calling the function within apply
candy_grouped3.count().groupby(level=0, group_keys=False).apply(sort_within_group)
```

```python
# Calling describe function on group3 object
candy_grouped4=candy_grouped3.apply(desc)
candy_grouped4
```

```python
# Unstacking the results from above step
candy_grouped4.unstack()
```

```python
# Creating a Groupby object on country name and Gender. This will split the data
candy_grouped5=candy_subset_df17.groupby(["CLEAN_COUNTRY_NAME","Q2: GENDER"])["Q3: AGE"]
# Performing apply and combine by calling the sort_within_group method on the grouped object
candy_grouped5.count().groupby(level=0, group_keys=False).apply(sort_within_group)
```

```python
# The above results can also be achieved using the lambda function
candy_grouped5.count().groupby(level=0, group_keys=False).apply(lambda x: x.sort_values(ascending=False))
```

### 11.1. Convert between string and date time

```python
# Importing the datetime
from datetime import datetime
# creating a sample date and converting between string and datetime types
date1=datetime(2023,5,6)
date1_str=str(date1)
type(date1), type(date1_str)
```

```python
# Creating a subset of MetObj dataframe and selecting rows where the date field are not null
metObj_subset_df1=metObj_orig_df[(~metObj_orig_df["Metadata Date"].isna()) & (~metObj_orig_df["Object Begin Date"].isna()) ]
# Setting an index for the dataframe
metObj_subset_df1=metObj_subset_df1.set_index(["Object ID"])
```

```python
# The dates in the met objects file are in various timestamp format. For simplicity, 
# the millisecond part is ignored in the datetime object
def split_date(str_date):
    """
    This function splits the Metadata Date based on the position of the period and ignores the 
    millisecond part as the data is inconsistent in the file.
    """
    return str_date[:-1].split(".")[0]
# Using the split_date function to ignore the millsecond part of the Metada Date
metObj_subset_df1["Metadata Date"]=metObj_subset_df1["Metadata Date"].apply(split_date)
# Converting the string to Datetime object
metObj_subset_df1["Metadata Date"]=pd.to_datetime(metObj_subset_df1["Metadata Date"])
```

```python
# Printing top 5 rows after date conversion
metObj_subset_df1.head()
```

```python
# Another way of converting to Date for the above step. Here the strptime is used to convert 
# the string to datetime object and then strftime is used to convert back to desired format.
# The downside of this approach is, it expects all  input data to be in the same specified format.
def convert_metadata_date(ts):
    # Converting String to Datetime object
    dt = datetime.strptime(ts[:-1], "%Y-%m-%dT%H:%M:%S.%f")
    # Converting the Date object to desired format
    formatted_timestamp = dt.strftime("%Y-%m-%d %H:%M:%S.%f")
    return formatted_timestamp
```

```python
# Converting Date to String:
def convert_date_to_string(input_time):
    """
    This function convers the Datetime object to desired string format using strftime method
    """
    return input_time.strftime("%Y-%m-%d %H:%M")
# Applying the funtion to Metadata date
metObj_subset_df1["Metadata Date"].apply(convert_date_to_string)
```

```python
# Converting string back to date
def convert_string_to_date(input_str):
    """
    This function converts string back to date using strptime method
    """
    return datetime.strptime(input_str,"%Y-%m-%d %H:%M")
# Applying the function to Metadata Date
metObj_subset_df1["Metadata Date"].apply(convert_date_to_string).apply(convert_string_to_date)
```
